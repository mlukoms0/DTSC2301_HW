#### Question 1

For the Ames Housing Data [<https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ames_house_prices.csv>], 
fit a linear model that predicts the *SalePrice* of a house from the *GarageArea*.   Evaluate whether or not
a linear model is appropriate for the relationship between these two features.


```python
ameshousedata = pd.read_csv('https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ames_house_prices.csv')
ameshousedata.head()
y = ameshousedata['SalePrice']
x = ameshousedata[['GarageArea']]
plt.scatter(x,y,s=5)
plt.xlabel("Garage Area")
plt.ylabel(" Sale Price")


model2 = LinearRegression()
model2.fit(x,y)


y_hat = model2.predict(x)
display = PredictionErrorDisplay(y_true=y,y_pred=y_hat)
display.plot()
rmse = root_mean_squared_error(y_true=y,y_pred=y_hat)
print(rmse)
```

##### ANSWERS FOR PROBELM 1
    The Linear Model seems to be a bad fit for these 2 features in this dataset. Doing a scatter plt shows us off the bat that there doesnt seem to be a even a moderate linear relationship. The leverage points as well as the residuals chart confirms for us that linear model is not approppriate.
